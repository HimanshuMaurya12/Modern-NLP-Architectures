Experiment 1
Write a Python program to Tokenization, POS tagging, and simple parsing

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag, CFG
from nltk.parse import RecursiveDescentParser

# Download necessary NLTK data (run once)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
# Added to resolve LookupError for English tagger
nltk.download('averaged_perceptron_tagger_eng')

def run_experiment_1():
    # --- Input Text ---
    text = "The quick brown fox jumps over the lazy dog. It is a sunny day."
    print(f"Original Text: {text}\n")

    # --- 1. Tokenization ---
    print("--- 1. Tokenization ---")

    # Sentence Tokenization
    sentences = sent_tokenize(text)
    print(f"Sentences: {sentences}")

    # Word Tokenization (using the first sentence as an example)
    words = word_tokenize(sentences[0])
    print(f"Words (Sentence 1): {words}\n")

    # --- 2. POS Tagging ---
    print("--- 2. POS Tagging ---")
    pos_tags = pos_tag(words)
    print("POS Tags (Word, Tag):")
    print(pos_tags)
    print("\n")

    # --- 3. Simple Parsing ---
    print("--- 3. Simple Parsing (Context-Free Grammar) ---")

    # Define a simple grammar for a specific sentence
    # Sentence: "The cat saw the dog"
    parse_sentence = "the cat saw the dog"
    parse_tokens = word_tokenize(parse_sentence)

    # Define a custom CFG to parse this specific sentence structure
    grammar = CFG.fromstring("""
      S -> NP VP
      VP -> V NP | V NP PP
      PP -> P NP
      V -> "saw" | "ate" | "walked"
      NP -> Det N | Det N PP
      Det -> "a" | "an" | "the"
      N -> "man" | "dog" | "cat" | "telescope" | "park"
      P -> "in" | "on" | "by" | "with"
    """)

    parser = RecursiveDescentParser(grammar)

    print(f"Parsing sentence: '{parse_sentence}'")
    try:
        for tree in parser.parse(parse_tokens):
            print("\nParse Tree:")
            tree.pretty_print() # Visualizes the tree structure
            # tree.draw() # Uncomment this to open a pop-up window with the tree
    except ValueError:
        print("No parse tree found for the given sentence and grammar.")

if __name__ == "__main__":
    run_experiment_1()
```
